<h1>Gathering and cleaning the data<h1>

Data received from: https://www.gov.uk/government/publications/court-martial-results-from-the-military-court-centres

```python
import pandas as pd
```

Load each dataframe and rename the columns to match.

```python
df1 = pd.read_excel('2010-2015.xls')
df1 = df1.drop(['Trial Court'], axis=1)
df1.head()
```

```python
df2 = pd.read_excel('2015-2016.xls', header=0)
df2 = df2.drop(['Trial Court'], axis=1)
df2 = df2.rename(columns={"Rank\rRank must be in full with no abbreviations": "Rank", "Charge (s) \rCharge(s) must be in full with no abbreviations": "Charge (s)"})
df2.head()
```

```python
df3 = pd.read_excel('2016-2017.xls', header=0)
df3 = df3.drop(['Column2', 'Trial Court', 'Unit'], axis=1)
df3 = df3.rename(columns={"Rank\nRank must be in full with no abbreviations": "Rank", "Charge (s) \nCharge(s) must be in full with no abbreviations": "Charge (s)", 'Total Sentence':'Sentence'})
df3.head()
```

```python
df4 = pd.read_excel('2017-2018.xls')
df4 = df4.drop(['Unit'], axis=1)
df4.head()
```

```python
df5 = pd.read_excel('2018.xls')
df5 = df5.drop(['Unit', 'Trial Court'], axis=1)
df5 = df5.rename(columns={'Overall Sentence':'Sentence', 'Charge': 'Charge (s)'})
df5.head()
```

```python
df6 = pd.read_excel('2018-2019.xls')
df6 = df6.rename(columns={'Overall Sentence':'Sentence','Rank\nRank must be in full with no abbreviations':'Rank','Charge (s) \nCharge(s) must be in full with no abbreviations':'Charge (s)'})
df6 = df6.drop(['Unit', 'Trial Court'], axis=1)
df6.head()
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6])
df.head()
```

```python
df.to_csv('court_martial_dataframe.csv', index = False)
```
